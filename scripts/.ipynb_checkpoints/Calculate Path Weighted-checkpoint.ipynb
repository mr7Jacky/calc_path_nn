## Load the model

```python
%matplotlib widget
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np
import pandas as pd
from pathcal.calculator import *
from pathcal.loader import *
from pathcal.layers import * 
from pathcal.helper import *
```

```python
exp = load_expriment()
```

```python
compression_rates = [4,10,20,30,40,50] # QMINIST, FASHION
# compression_rates = [2,4,10,20,50] # CIFAR
models = [i for i in range(10)]
classes = [i for i in range(10)]
```

## Calculate actual paths

```python
paths = {}
isPos = False
blockPrint()
for c in compression_rates:
    temp_path = torch.zeros((len(models), 10))
    for m in range(len(models)):
        load_model(exp, model=m, compression=c, stg="finetuned")
        prev_paths = None
        prev_layer = None
        last_num = 0
        count = 0
        for layer in exp.model.modules():
            last_num += 1
        for layer in exp.model.modules():
            cur_path = []
            if isConv(layer):
                weights = list(layer.named_parameters())[0][1]
                cur_path = cal_conv_paths_weighted(prev_paths, weights, pos=isPos)
                prev_layer = layer
                prev_paths = cur_path
            elif isLinear(layer):
                weights = list(layer.named_parameters())[0][1]
                if isConv(prev_layer):
                    prev_paths = expend_paths(prev_paths, weights)
                if last_num == count + 1:
                    cur_paths = cal_linear_paths_weighted(prev_paths, weights, 
                                                          0.1, pos=isPos)
                else:
                    cur_paths = cal_linear_paths_weighted(prev_paths, weights, 
                                                          pos=isPos)
                prev_layer = layer
                prev_paths = cur_paths
            count += 1
        temp_path[m] = prev_paths
    df = pd.DataFrame(temp_path.detach().numpy(), index=[f'model{m}' for m in models], columns=[f'class{c}' for c in classes])
    df.to_csv(f"./qminist/weighted_path_r{c}.csv")
enablePrint()
```
