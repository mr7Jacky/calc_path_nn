## Load the model

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathcal.calculator import *
from pathcal.loader import *
from pathcal.layers import * 
from pathcal.helper import *
```

```python
exp = load_expriment()
```

```python
compression_rates = [4,10,20,30,40,50] # QMINIST, FASHION
# compression_rates = [2,4,10,20,50] # CIFAR
models = [i for i in range(10)]
classes = [i for i in range(10)]
```

```python
flows = {}
blockPrint()
for c in compression_rates:
    temp_path = torch.zeros((len(models), 10))
    for m in range(len(models)):
        # Load the model
        load_model(exp, model=m, compression=c, stg="pruning")
        # keep a records of previous flow for calculation purpose
        prev_flows = None
        # Keep a records of previous layer type for detecting transition between conv and linear
        prev_layer = None
        for layer in exp.model.modules():
            cur_path = []
            # Calculate the layer based on layer types
            # Only consider conv and linear currently
            if isConv(layer):
                # Get current weights
                weights = list(layer.named_parameters())[0][1]
                cur_path = cal_conv_flows(prev_flows, weights)
                prev_layer = layer
                prev_flows = cur_path
            elif isLinear(layer):
                # Get current weights
                weights = list(layer.named_parameters())[0][1]
                # I fprev layer is conv requires reshape
                if isConv(prev_layer):
                    prev_flows = expend_flows(prev_flows, weights)
                cur_flows = cal_linear_flows(prev_flows, weights)
                prev_layer = layer
                prev_flows = cur_flows
        temp_path[m] = prev_flows # save the flow of output layer
    df = pd.DataFrame(temp_path.detach().numpy(), index=[f'model{m}' for m in models], columns=[f'class{c}' for c in classes])
    df.to_csv(f"./fashion/flow_r{c}.csv")
enablePrint()
```

## Calculate actual flows

```python
flows = {}

for c in compression_rates:
    temp_path = torch.zeros((len(models), 10))
    for m in range(len(models)):
        # Load the model
        load_model(exp, model=m, compression=c, stg="pruning")
        # keep a records of previous flow for calculation purpose
        prev_flows = None
        # Keep a records of previous layer type for detecting transition between conv and linear
        prev_layer = None
        for layer in exp.model.modules():
            cur_path = []
            # Calculate the layer based on layer types
            # Only consider conv and linear currently
            if isConv(layer):
                # Get current weights
                weights = list(layer.named_parameters())[0][1]
                cur_path = cal_conv_flows(prev_flows, weights)
                prev_layer = layer
                prev_flows = cur_path
            elif isLinear(layer):
                # Get current weights
                weights = list(layer.named_parameters())[0][1]
                # I fprev layer is conv requires reshape
                if isConv(prev_layer):
                    prev_flows = expend_flows(prev_flows, weights)
                cur_flows = cal_linear_flows(prev_flows, weights)
                prev_layer = layer
                prev_flows = cur_flows
        temp_path[m] = prev_flows # save the flow of output layer
    flows[c] = temp_path.mean(axis=0).detach().numpy() # Take avg of all models
```

```python
pd.DataFrame.from_dict(flows, orient='index', columns=[f'class{c}' for c in range(10)])
```

```python
plt.figure()
classes = [i for i in range(10)]
for key, value in flows.items():
    ret = value
    plt.plot(classes, ret.detach().numpy(), label=key)
plt.legend()
```
